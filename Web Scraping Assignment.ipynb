2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
and make data frame

```python
page=requests.get('https://www.imdb.com/list/ls091520106/')
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
movietitle=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    movietitle.append(i.a.text.replace("\n",""))
movietitle
```

```python
rating=[]
for i in soup.find_all('div',class_="ipl-rating-star small"):
    rating.append(i.text.replace("\n",""))
print(len(rating))
print("rating of all the titles:",rating)
```

```python
release=[]
year=soup.find_all('span',class_="lister-item-year text-muted unbold")
for i in year:
    i=i.get_text().replace("\n",'')
    i=i.strip("()")
    release.append(i)
print(len(release))
print("year of release :",release)
```

```python
import pandas as pd
df=pd.DataFrame()
df['Movie Title']=movietitle
df["Year of Release"]=release
df["Rating"]=rating
df
```
